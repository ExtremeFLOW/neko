```python
%matplotlib notebook
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("TSwave_cheb_2D.bin", mode='rb') as file:
    ss2d = np.fromfile(file)
    ss2d = ss2d.reshape((7, 161))
    ss2d = ss2d.T
with open("TSwave_cheb_3D.bin", mode='rb') as file:
    ss3d = np.fromfile(file)
    ss3d = ss3d.reshape((7, 161))
    ss3d = ss3d.T
```

```python
# 2D mode: alpha = 1.12, beta = 0
fig=plt.figure()
plt.plot(ss2d[:,0],ss2d[:,1],'r-')
plt.plot(ss2d[:,0],ss2d[:,2],'r--')
plt.plot(ss2d[:,0],ss2d[:,3],'b-')
plt.plot(ss2d[:,0],ss2d[:,4],'b--')
plt.xlabel(r'$y$');plt.ylabel(r'u,v,w')
plt.show()
```

```python
# 3D mode: alpha = 1.12, beta = 2.1
fig=plt.figure()
plt.plot(ss3d[:,0],ss3d[:,1],'r-')
plt.plot(ss3d[:,0],ss3d[:,2],'r--')
plt.plot(ss3d[:,0],ss3d[:,3],'b-')
plt.plot(ss3d[:,0],ss3d[:,4],'b--')
plt.plot(ss3d[:,0],ss3d[:,5],'g-')       # negative for -beta mode
plt.plot(ss3d[:,0],ss3d[:,6],'g--')      # negative for -beta mode
plt.xlabel(r'$y$');plt.ylabel(r'u,v,w')
plt.show()
```

```python
# 3D mode: alpha = 1.12, beta = -2.1
fig=plt.figure()
plt.plot(ss3d[:,0],ss3d[:,1],'r-')
plt.plot(ss3d[:,0],ss3d[:,2],'r--')
plt.plot(ss3d[:,0],ss3d[:,3],'b-')
plt.plot(ss3d[:,0],ss3d[:,4],'b--')
plt.plot(ss3d[:,0],-ss3d[:,5],'g-')       # negative for -beta mode
plt.plot(ss3d[:,0],-ss3d[:,6],'g--')      # negative for -beta mode
plt.xlabel(r'$y$');plt.ylabel(r'u,v,w')
plt.show()
```
